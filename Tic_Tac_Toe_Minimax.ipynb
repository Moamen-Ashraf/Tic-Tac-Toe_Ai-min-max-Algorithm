# Tic-Tac-Toe — Minimax AI

A Python implementation of Tic-Tac-Toe where the computer plays optimally using the **Minimax algorithm**.

- **Human** plays as `X`
- **Computer** plays as `O`

### Board Layout
Positions you can enter (1–9):
```
 1 | 2 | 3
-----------
 4 | 5 | 6
-----------
 7 | 8 | 9
```

## Board Initialization

````python
from IPython.display import clear_output

# Global board: flat list of 9 cells, each ' ', 'X', or 'O'
# Internally indexed 0-8, mapped from user input 1-9
board = [" "] * 9
````

## Helper Functions

````python
def is_Win(player):
    """
    Check if the given player ('X' or 'O') has won.
    Checks all 8 winning combinations: 3 rows, 3 columns, 2 diagonals.
    """
    # Rows: [0,1,2], [3,4,5], [6,7,8]
    for i in range(3):
        if board[i*3] == board[i*3 + 1] == board[i*3 + 2] == player:
            return True

    # Columns: [0,3,6], [1,4,7], [2,5,8]
    for i in range(3):
        if board[i] == board[i+3] == board[i+6] == player:
            return True

    # Diagonals: [0,4,8] and [2,4,6]
    if board[0] == board[4] == board[8] == player:
        return True
    if board[2] == board[4] == board[6] == player:
        return True

    return False


def display_board():
    """Print the current board state as a formatted 3x3 grid."""
    print('-' * 13)
    for i in range(3):
        print("|", board[i*3], "|", board[i*3 + 1], "|", board[i*3 + 2], "|")
        print('-' * 13)
````

## Minimax Algorithm

The core AI logic. Recursively scores every possible future game state:

| Outcome | Score |
|---------|-------|
| AI (O) wins | +1 |
| Human (X) wins | -1 |
| Draw | 0 |

- **Maximizing** (AI's turn): picks the move with the **highest** score
- **Minimizing** (Human's turn): picks the move with the **lowest** score

````python
def minimax(board, depth, isMaximizing):
    """
    Minimax recursive function.

    Parameters:
        board         : current board state (modified in-place, then restored)
        depth         : current recursion depth
        isMaximizing  : True = AI turn (O), False = Human turn (X)

    Returns:
        +1 if O wins, -1 if X wins, 0 for draw
    """
    # Base cases — terminal states
    if is_Win('X'):
        return -1   # Human wins → bad for AI
    elif is_Win('O'):
        return 1    # AI wins → good for AI
    elif ' ' not in board:
        return 0    # Draw

    if isMaximizing:
        # AI's turn: try all empty cells, pick highest score
        best_score = float('-inf')
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'                           # try move
                score = minimax(board, depth + 1, False) # recurse as minimizer
                board[i] = ' '                           # undo move
                best_score = max(score, best_score)
        return best_score
    else:
        # Human's turn: try all empty cells, pick lowest score
        best_score = float('inf')
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'                          # try move
                score = minimax(board, depth + 1, True) # recurse as maximizer
                board[i] = ' '                          # undo move
                best_score = min(score, best_score)
        return best_score


def min_max(board):
    """
    AI move selector: evaluates all available moves using minimax,
    then applies the best one to the board.
    """
    final_score = float('-inf')
    final_i = None

    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'                    # try move
            score = minimax(board, 0, False)   # score this future
            board[i] = ' '                    # undo move

            if score > final_score:
                final_score = score
                final_i = i                   # track best move index

    board[final_i] = 'O'  # apply best move
````

## Game Loop

````python
def Play_XO():
    """
    Main game loop.
    - Human (X) enters a position 1-9
    - AI (O) responds with the optimal Minimax move
    - Checks for win or draw after every move
    """
    player = 'X'  # Human always goes first

    while True:
        display_board()

        if player == 'X':
            # ── Human turn ────────────────────────────────────────────────
            while True:
                position = int(input("Player " + player + ", enter the position (1-9): "))

                if position < 1 or position > 9:
                    print('\nInvalid position, please enter a position between 1 and 9\n')
                    continue
                elif board[position - 1] != ' ':
                    print(f'\nInvalid position, cell is occupied by {board[position - 1]}\n')
                else:
                    break

            board[position - 1] = player

            if is_Win(player):
                display_board()
                print("Player " + player + " Wins!")
                break
            elif ' ' not in board:
                print("It's a Tie!")
                break

        else:
            # ── AI turn ───────────────────────────────────────────────────
            min_max(board)  # AI picks and applies best move

            if is_Win('O'):
                display_board()
                print("Computer Wins!")
                break
            elif ' ' not in board:
                print("It's a Tie!")
                break

        # Switch player and clear screen
        player = 'O' if player == 'X' else 'X'
        clear_output(wait=True)


# Start the game
Play_XO()
````
